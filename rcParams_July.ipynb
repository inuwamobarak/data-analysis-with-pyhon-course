```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib as mpl
```

```python
df = pd.read_csv("/content/zomato.csv")
df.head()
```

```python
plt.plot(df["rate (out of 5)"], df['num of ratings'])
plt.show()
```

```python
# Global Configuration

mpl.rcParams['lines.linewidth'] = 4
```

```python
plt.plot(df["rate (out of 5)"], df['num of ratings'])
plt.show()
```

```python
mpl.rcParams['font.size'] = 16
```

```python
plt.plot(df["rate (out of 5)"], df['num of ratings'])
plt.show()
```

```python
# resetting to defaults

mpl.rcdefaults()
```

```python
plt.plot(df["rate (out of 5)"], df['num of ratings'])
plt.show()
```

```python
# common parameters

mpl.rcParams['axes.labelsize'] = 25
mpl.rcParams['legend.loc'] = 'upper left'
mpl.rcParams['xtick.labelsize'] = 6
mpl.rcParams['ytick.labelsize'] = 10
```

```python
plt.plot(df['num of ratings'])
plt.title('Ratings')
plt.legend(['num of ratings'])
plt.show()
```

```python
mpl.rcdefaults()
```

```python
with plt.rc_context({'lines.linewidth': 3, 'font.family': 'serif'}):
  plt.plot(df['num of ratings'])
  plt.title('Ratings')
  plt.legend(['num of ratings'])
  plt.show()
```
