```python
import plotly.express as px

df = px.data.gapminder()
df.head()
```

```python
fig = px.line(df, x='year', y='gdpPercap', color='country', title='GDP Per Capital Over Years')
fig.show()
```

```python
# Scatter Plot

df = px.data.iris()
df.columns
```

```python
fig = px.scatter(df, x='sepal_width', y='sepal_length', color='species_id', title='The Iris Flower')
fig.show()
```

```python
# Bar Chart
df = px.data.tips()
df.columns
```

```python
fig = px.bar(df, x='day', y='total_bill', color='sex', facet_col='smoker', title='Total Bill by Day')
fig.show()
```

```python
# Pie Chart
fig = px.pie(df, names='day', title='Distribution by Day')
fig.show()
```

```python
# Histogram

fig = px.histogram(df, x='total_bill', color='time', title='Total Bill')
fig.show()
```

```python
# Box Plot

fig = px.box(df, x='day', y='total_bill', color='time', title='Total Bill Box Plot')
fig.show()
```

```python
# Violin Plot
fig = px.violin(df, x='day', y='total_bill', color='sex', title='Total Bill Violin Plot')
fig.show()
```
