```python
# Read smartphone dataset as smartphone_df

smartphone_df = pd.read_csv("/content/smartphones.csv")
smartphone_df.head()
```

```python
smartphone_df.info()
```

```python
# smartphone_df = smartphone_df.dropna()
# smartphone_df.info()
```

```python
smartphone_df = smartphone_df.dropna(subset=['processor_speed'])
smartphone_df.info()
```

```python
# Filling Missing Values

## Filling with mean()
smartphone_df['avg_rating'] = smartphone_df['avg_rating'].fillna(smartphone_df['avg_rating'].mean())
```

```python
smartphone_df.info()
```

```python
smartphone_df.shape
```

```python
smartphone_df['processor_brand'] = smartphone_df['processor_brand'].fillna('Unknown brand')
smartphone_df.info()
```

```python
## Handling Duplicates

duplicates = smartphone_df.duplicated()
duplicates.sum()
```

```python
### Removing Duplicates

no_duplicates = smartphone_df.drop_duplicates()
```

```python
model_duplicates = smartphone_df.duplicated(subset=['model'])
model_duplicates.sum()
```

```python
## Using Lambda

f = lambda x: x * 2
smartphone_df['resolution_height'] = smartphone_df['resolution_height'].apply(f)
```
