## Checking for Missing Values

```python
# Sample Dataset
df = pd.read_csv('/content/drive/MyDrive/LOCTEC/DAWP 2023/Dataset2/Smartphone/smartphones.csv')
df.head()
```

```python
# Check missing values
df.isnull()
```

```python
# Check missing values
df.notnull().sum()
```

## Remove Rows with Missing Values

```python
drop_rows = df.dropna()
drop_rows
```

```python
# Dropping rows with NaN columns
drop_rows_with_cols = df.dropna(subset=['battery_capacity'])

drop_rows_with_cols
```

### Filling Missing Values

```python
# Filling with specific value
fill_value = df.fillna(2)

fill_value
```

```python
# Filling with mean
df['battery_capacity'].fillna(df['battery_capacity'].mean(), inplace=True)
df
```

```python
# Filling with median
df['battery_capacity'].fillna(df['battery_capacity'].median(), inplace=True)
df
```

## Handling Duplicates

```python
# Checking for duplicates
duplicate_df = df.duplicated().sum()
duplicate_df
```

```python
# Checking for duplicates
duplicate_df = df.avg_rating.duplicated().sum()
print("Total duplicated ratings:", duplicate_df)
```

```python
# Removing Duplicates
removed_dup_rows = df.drop_duplicates()
removed_dup_rows
```

```python
# Removing Duplicates
removed_dup_rows = df.drop_duplicates(subset=['avg_rating'])
removed_dup_rows
```

```python
# Checking for duplicates
duplicate_df = removed_dup_rows.avg_rating.duplicated().sum()
print("Total duplicated ratings:", duplicate_df)
```
