## SQL for Data Analysis

```python
# Importing Dependencies
import sqlite3

# Connect to a Database
conn = sqlite3.connect(':memory:')
```

:memory: inplies an in memory database

```python
# Connection Object
c = conn.cursor()

# Query
c.execute("""
  CREATE TABLE sales
  (date text, productID text, productname text, quantity real)
""")
```

```python
# Inserting into our Table
c.execute("INSERT INTO sales VALUES('19-09-2023','AB123', 'Rice', 5)")

# Commit the transaction to db
conn.commit()
```

```python
# Querying our db
c.execute("SELECT * FROM sales")
print(c.fetchall())
```

```python
# Converting to DataFrame
import pandas as pd

df = pd.read_sql_query("SELECT * FROM sales", conn)
```

## ORDER BY Clause(For Sorting)

```python
# ORDER BY

age_order = pd.read_sql("""
        SELECT * FROM df
        ORDER BY Age
        """, conn)

age_order.head()
```

```python
# ORDER BY(ASC/DESC)

age_order = pd.read_sql("""
        SELECT * FROM df
        ORDER BY Age DESC
        """, conn)

age_order.head()
```

## Group By Clause

GROUP BY Clause groups rows that have same value in specified columns. Works with SUM, AVG, etc.

```python
 # GROUP BY(ASC/DESC)

gender_group = pd.read_sql("""
        SELECT Gender, AVG(Spending_Score) FROM df
        GROUP BY Gender
        """, conn)

gender_group.head()
```

## LIMIT, OFFSET, and DISTINCT

```python
# LIMIT
limit = pd.read_sql("""
        SELECT * FROM df
        LIMIT 7
        """, conn)

limit
```

```python
# OFFSET
offset = pd.read_sql("""
        SELECT * FROM df
        LIMIT 7
        OFFSET 5
        """, conn)

offset
```

```python
# DISTINCT
distinct = pd.read_sql("""
        SELECT DISTINCT Age FROM df
        LIMIT 7
        OFFSET 5
        """, conn)

distinct
```

## Renaming Columns

## SQL Functions

```python
# UPPER (Convert text to uppercase)
upper = pd.read_sql("""
        SELECT UPPER(Gender) as "Sex",
        Spending_Score as Score FROM df
        LIMIT 10
        OFFSET 10
        """, conn)

upper
```

```python
# LOWER (Convert text to lowercase)
lower = pd.read_sql("""
        SELECT LOWER(Gender) as "Sex",
        Spending_Score as Score FROM df
        LIMIT 10
        OFFSET 10
        """, conn)

lower
```

```python
# Lenght (Returns the number of strings in a text)
lenght = pd.read_sql("""
        SELECT Length(Gender) as "Sex",
        Spending_Score as Score FROM df
        LIMIT 10
        OFFSET 10
        """, conn)

lenght
```

```python
# SUBSTRING (Returns a part of a string in a text)
substring = pd.read_sql("""
        SELECT SUBSTRING(Gender, 2) as "Sex",
        Spending_Score as Score FROM df
        LIMIT 10
        OFFSET 10
        """, conn)

substring
```

```python
df2 = pd.read_sql("""
        SELECT CustomerID,
        Gender as Sex,	Age,
        "Annual_Income_(k$)",
        "Spending_Score" FROM df
        """, conn)

df2
```

```python
# MIN
min_df = pd.read_sql("""
        SELECT MIN(Age) FROM df
             """, conn)

min_df
```

```python
# MAX
max_df = pd.read_sql("""
        SELECT MAX(Age) as max_age FROM df
             """, conn)

max_df
```

```python
# AVG
avg_df = pd.read_sql("""
        SELECT AVG(Age) FROM df
             """, conn)

avg_df
```

## Boolean Expressions and Concatenation

```python
# Boolean Expression

bool_df = pd.read_sql("""
    SELECT * FROM df
    WHERE(Age>45 OR Spending_Score>35)""", conn)

bool_df.head()
```

```python
# Boolean Expression

bool_df = pd.read_sql("""
    SELECT * FROM df
    WHERE(Age>45 OR Spending_Score>35)
    AND NOT (Gender='Male')
    """, conn)

bool_df.head()
```

```python
# Boolean Expression

bool_df = pd.read_sql("""
    SELECT Age || Gender FROM df
    """, conn)

bool_df.head()
```

```python
# Concatenate

bool_df = pd.read_sql("""
    SELECT Age ||' '|| Gender FROM df
    """, conn)

bool_df.head()
```

```python
# Using CASE Clause

case_df = pd.read_sql("""
    SELECT Age,
      CASE
        WHEN Spending_Score > 30 THEN 'A'
        WHEN "Annual_Income_(k$)" > 7 THEN 'B'
        ELSE 'C'
      END AS 'Target'
    FROM df
    """, conn)

case_df.head()
```

# Creating Views and Materialized Views
