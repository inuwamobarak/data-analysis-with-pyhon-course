The above column names indicate the first two columns are not useful. We proceed by dropping them.

```python
df = df.drop(['Unnamed: 0.1', 'Unnamed: 0'], axis=1)
```

```python
df.columns
```

```python
# Replacing #,'@ with nothing
df['restaurant name'] = df['restaurant name'].str.replace('#', '')
df['restaurant name'] = df['restaurant name'].str.replace("'@ ", '')
df['restaurant name'] = df['restaurant name'].str.replace("'@", '')

df['area'] = df['area'].str.replace(",", ', ')
```

```python
df.head()
```

```python
df['area'][1] = 'Nigeria'
```

```python
df.head()
```

```python
# Renaming
df = df.rename(columns = {'restaurant name':'restaurant_name',
                     'restaurant type':'restaurant_type',
                     'num of ratings' : 'num_of_ratings',
                     "rate (out of 5)" : "ratting_over_5",
                     'avg cost (two people)':'avg_cost_two_people',
                     'table booking' : 'table_booking',
                     'cuisines type' : 'cuisines_type',
                     'local address' : 'local_address'
                     })
df.columns
```
